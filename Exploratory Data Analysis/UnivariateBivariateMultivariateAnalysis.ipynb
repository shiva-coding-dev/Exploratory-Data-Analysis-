# Univariate Analysis

```python
df_setosa = df.loc[df['species']=='setosa']
df_versicolor = df.loc[df['species'] == 'versicolor']
df_virginica = df.loc[df['species'] == 'virginica']
```

```python
plt.plot(df_setosa['sepal_length'] , np.zeros_like(df_setosa['sepal_length']) , 'x')
plt.plot(df_virginica['sepal_length'] , np.zeros_like(df_virginica['sepal_length']) , 'x')
plt.plot(df_versicolor['sepal_length'] , np.zeros_like(df_versicolor['sepal_length']) , 'x')
plt.xlabel('Sepal Length')
plt.show()
```

# Bivariate Analysis

```python
sns.FacetGrid(df , hue='species' , height = 5).map(plt.scatter , "petal_length" , "sepal_width").add_legend()
plt.show()
```

Observation - setosa has petal length less than 3 always 

# Multivariate Analysis 

```python
sns.pairplot( data = df , hue='species', palette='rainbow')
```

Observation 1 Petal length <=2 and Petal Width <=1  -> Setosa  

Observation 2- [1<= Petal width <=1.7]  and [5<= Sepal length <=7]  -> Versicolor
