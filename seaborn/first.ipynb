### Correlation 
* Correlation means how a variable is changing with respect to other variable 
* Three types of correlation 
* * Positive
* * Negative
* * Zero
  

```python

data = {
    'hours_studied': [1, 2, 3, 4],
    'marks_scored': [10, 20, 30, 40]
}

df = pd.DataFrame(data)
print(df.corr())
```

### Correlation doesn't mean Casuation 
* Means if A is correlated with the B then it does not mean that by changing A , B will get changed

# JoinPlot 
### A JoinPlot allows to study relation between 2 numeric variables. The Central chart display their correlation. It is usually a scatterplot , a hexbin plot 

### Univariate analysis

```python
sns.jointplot(x = 'tip' , y = 'total_bill', data = df, kind='hex')
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill', data = df, kind='reg')
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill', data = df, kind='kde')
```

# Pair Plot
### seaborn.pairplot() creates a matrix of plots showing pairwise relationships between all numeric variables in a DataFrame.

It’s a quick way to explore how features relate to one another visually.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.displot(df['tip'])
```
