### Working with DATA

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Data structure = [restaurant_id ,2021 , 2022 , 2023 , 2024 ]
sale_data = np.array([[1 , 150000 , 180000 , 220000 , 400000], #Paradise Biryani 
                      [2 , 120000 , 150000 , 230000 , 340000], #Beijing foods
                      [3 , 140000 , 290000 , 340000 , 550000], #ramdhani hotels
                      [4 , 110000 , 235000 , 455000 , 505000],  #gharJaisa
                      [5 , 115000 , 232000 , 475000 , 605000]  #foottech
                      ])

print("Zomato Sale Analysis :");
print("Shape of the Data :", sale_data.shape)
min_sales = np.min(sale_data[:,1:] , axis = 1)
col_idx = np.argmin(sale_data[:,1:] , axis = 1)

for i in range(len(sale_data)):
    print(f"Minimum Sales of Company having id {i+1} is : {min_sales[i]} in the year {col_idx[i]+2021}")

max_sales = np.max(sale_data[:,1:] , axis =1)
max_col_idx = np.argmax(sale_data[:,1:],axis = 1)
for i in range(len(sale_data)):
    print(f"Maxmimum Sales of the Company {i+1} is {max_sales[i]} in the year {max_col_idx[i]+2021}")

print("Average Sale per Restaurant ")
```

### Finding Average Sales 

```python
import numpy as np 
sale_data = np.array([[1 , 150000 , 180000 , 220000 , 400000], #Paradise Biryani 
                      [2 , 120000 , 150000 , 230000 , 340000], #Beijing foods
                      [3 , 140000 , 290000 , 340000 , 550000], #ramdhani hotels
                      [4 , 110000 , 235000 , 455000 , 505000],  #gharJaisa
                      [5 , 115000 , 232000 , 475000 , 605000]  #foottech
                      ])
avg_sale = np.mean(sale_data[:,1:], axis = 1)
print(avg_sale)
```

### Cumulative Sales

```python
cumsum = np.cumsum(sale_data[:,1:], axis = 1)
print(cumsum)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum , axis = 0))
plt.title("Average Cummulative Sales across all Restaurants ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("Vector addition :", vector1+vector2)
print("Vector multiplicaton" , vector1*vector2)
print("Dot product :", np.dot(vector1 , vector2))
dotPr  = np.dot(vector1 , vector2)
mod1 = np.linalg.norm(vector1)
mod2 = np.linalg.norm(vector2)
angle = np.arccos(dotPr/((mod1)*(mod2)))
print("Angles between vectors :", angle)
```
